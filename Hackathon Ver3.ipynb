```python
model = torch.hub.load('ultralytics/yolov5', 'yolov5s', pretrained=True)
```

```python
tracker = DeepSort(max_age=30, n_init=3, max_cosine_distance=0.4)
```

```python
def process_video(video_path):
    cap = cv2.VideoCapture(video_path)
    results_data = []
    frame_num = 0

    while cap.isOpened():
        ret, frame = cap.read()
        if not ret:
            break

        frame_num += 1
        print(f"Processing frame {frame_num}")

        # Run YOLOv5 detection
        results = model(frame)
        detections = results.pandas().xyxy[0]

        print("Raw Detections:")
        print(detections[['name', 'confidence']].head(10))

        if detections.empty:
            print("No detections on this frame.")
            continue

        print("Detected classes:", detections['name'].unique())

        dets = []
        for _, row in detections.iterrows():
            bbox = [int(row['xmin']), int(row['ymin']), int(row['xmax']), int(row['ymax'])]
            conf = row['confidence']
            class_name = row['name']
            dets.append(([bbox[0], bbox[1], bbox[2] - bbox[0], bbox[3] - bbox[1]], conf, class_name))

        if not dets:
            print("No valid detections to track.")
            continue

        tracks = tracker.update_tracks(dets, frame=frame)

        for track in tracks:
            if not track.is_confirmed():
                continue

    
            track_id = track.track_id
            ltrb = track.to_ltrb()
            class_name = track.get_det_class()

            results_data.append({
                'timestamp': datetime.now().strftime("%Y-%m-%d %H:%M:%S"),
                'track_id': track_id,
                'vehicle_type': class_name,
                'bbox': ltrb
            })

            # Annotate frame
            x1, y1, x2, y2 = map(int, ltrb)
            cv2.rectangle(frame, (x1, y1), (x2, y2), (0, 255, 0), 2)
            cv2.putText(frame, f'ID: {track_id}', (x1, y1 - 10),
                        cv2.FONT_HERSHEY_SIMPLEX, 0.6, (255, 255, 0), 2)

    cap.release()
    return pd.DataFrame(results_data)
```

```python
ef generate_report(df):
    if df.empty:
        print("No data collected. Please check the video or model setup.")
        return

    df.to_csv("traffic_report.csv", index=False)

    plt.figure(figsize=(12, 6))
    sns.countplot(data=df, x='vehicle_type')
    plt.title("Vehicle Type Distribution")
    plt.xlabel("Vehicle Type")
    plt.ylabel("Count")
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.show()

    print(f"Total Unique Vehicles Tracked: {df['track_id'].nunique()}")
```

```python
video_path = "traffic2.mp4"
df = process_video(video_path)
print(df.head())
generate_report(df)
```
